Task 1

```python
data2 = {
    'Date': pd.date_range(start='2023-01-01', periods=10),
    'Product_A': [120, 150, 130, 110, 140, 160, 135, 125, 145, 155],
    'Product_B': [90, 110, 100, 80, 95, 105, 98, 88, 102, 112],
    'Product_C': [75, 80, 85, 70, 88, 92, 78, 82, 87, 90]
}

df2 = pd.DataFrame(data2)
df2
```

```python
df2['Total'] = df2[['Product_A', 'Product_B', 'Product_C']].sum(axis=1)
df2
```

```python
max_val = df2[['Product_A', 'Product_B', 'Product_C']].sum(axis=1).max()
df2[df2['Total'] == max_val]
```

```python
# df2['Change(%)'] = df2[]
percentage_change = df2[['Product_A', 'Product_B', 'Product_C']].pct_change() * 100

# Add Date column for reference
percentage_change.columns = ['Product_A(%)', 'Product_B(%)', 'Product_C(%)']
percentage_change['Date'] = df2['Date']
percentage_change['Product_A'] = df2['Product_A']
percentage_change['Product_B'] = df2['Product_B']
percentage_change['Product_C'] = df2['Product_C']

print(percentage_change)
```

```python
plt.plot( df2['Product_A'], df2['Date'], color='red', label='Product A')
plt.plot( df2['Product_B'], df2['Date'], label='Product B')
plt.plot( df2['Product_C'], df2['Date'], c='g', label='Product C')
plt.xlabel('Price')
plt.ylabel('Date')
plt.title('Daily analysis of products')

plt.legend()

plt.show()
```
